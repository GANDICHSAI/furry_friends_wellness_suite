```python
import random

first_names = ["James", "John", "Robert", "Michael", "William", "David", "Richard", "Joseph", "Thomas", "Charles", "Christopher"]
last_names = ["Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller", "Davis", "Rodriguez", "Martinez"]

customers = []
for i in range(1007, 1050):  # generate customer with id from 1007 to 1050, order randomized
    first = random.choice(first_names)
    last = random.choice(last_names)
    email = f"{first.lower()}.{last.lower()}@example.com"
    password = "password123"
    customers.append((first, last, email, password))

# Now, format this list into an SQL query
values = ',\n'.join([f"('{f}', '{l}', '{e}', '{p}')" for f, l, e, p in customers])
insert_query = f"INSERT INTO Customer (first_name, last_name, email, password) VALUES \n{values};"
print(insert_query)
```

```python
pet_names = ["Buddy", "Lucy", "Charlie", "Bella", "Max", "Luna"]
types = ["Dog", "Cat"]
colors = ["Brown", "Black", "White", "Golden", "Grey", "Mixed"]
genders = ["Male", "Female"]

pets = []
for cust_id in range(1007, 1050): # for randomly generated customers above, assign random pet
    name = random.choice(pet_names)
    age = random.randint(1, 10)  # Pet age between 1 and 10
    color = random.choice(colors)
    gender = random.choice(genders)
    weight = random.uniform(5.0, 15.0)  # Weight between 5.0 to 15.0 pounds
    pet_type = random.choice(types)
    pets.append((cust_id, name, age, color, gender, round(weight, 1), pet_type))

# Format this list into an SQL query
values = ',\n'.join([f"({cid}, '{n}', {a}, '{c}', '{g}', {w}, '{t}')" for cid, n, a, c, g, w, t in pets])
insert_query = f"INSERT INTO Pet (customer_id, pet_name, age, color, gender, weight, type) VALUES \n{values};"
print(insert_query)
```

```python
import random
from datetime import timedelta, datetime

def random_date(start, end):
    """Generate a random date between `start` and `end`"""
    return (start + timedelta(
        days=random.randint(0, int((end - start).days)),
        hours=random.randint(0, 23),
        minutes=random.randint(0, 59)
    )).strftime('%Y-%m-%d %H:%M:%S')

start_date = datetime.strptime('2023-01-01', '%Y-%m-%d')
end_date = datetime.strptime('2023-12-31', '%Y-%m-%d')
statuses = ['Pending', 'Completed', 'Cancelled']
store_ids = [2001, 2002, 2003]  # randomly generate across 3 store IDs
service_ids = [7001, 7002, 7003]  # randomly generate accrose 3 service packages

appointments = []
for cust_id in range(1007, 1050):  # Customer IDs from 1007 to 1049
    for _ in range(random.randint(1, 3)):  # Each customer makes 1 to 3 appointments
        pet_id = random.randint(4007, 4049)  # Randomize pet ID from available range in db
        date = random_date(start_date, end_date)
        status = random.choice(statuses)
        store_id = random.choice(store_ids)
        service_id = random.choice(service_ids)
        appointments.append((cust_id, store_id, service_id, pet_id, date, status))

# Format this list into an SQL query
values = ',\n'.join([f"({c}, {s}, {sv}, {p}, '{d}', '{st}')" for c, s, sv, p, d, st in appointments])
insert_query = f"INSERT INTO Appointment (cust_id, store_id, service_id, pet_id, date, status) VALUES \n{values};"
print(insert_query)
```

```python
import matplotlib.pyplot as plt

# Define service mapping based on the provided codes
service_mapping = {7001: 'BATH AND FULL HAIRCUT', 7002: 'BATH AND BRUSH', 7003: 'BATH ONLY'}

# Filter data for the relevant services
relevant_data = app_data[app_data['service_id'].isin(service_mapping.keys())]

# Count the frequency of each service
service_counts = relevant_data['service_id'].map(service_mapping).value_counts().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
service_counts.plot(kind='bar', color='skyblue', alpha=0.7)
plt.title('Frequency of Each Service Type')
plt.xlabel('Service Type')
plt.ylabel('Number of Appointments')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Service Popularity

```python
# Convert 'date' to datetime format
relevant_data['date'] = pd.to_datetime(relevant_data['date'])

# Extract year and month from the date for grouping
relevant_data['year_month'] = relevant_data['date'].dt.to_period('M')

# Group by year-month and service_id, then count frequencies
monthly_trends = relevant_data.groupby(['year_month', 'service_id']).size().unstack(fill_value=0)
monthly_trends.columns = [service_mapping[id] for id in monthly_trends.columns]  # Rename columns to service names


plt.figure(figsize=(14, 7))
for service in monthly_trends.columns:
    plt.plot(monthly_trends.index.astype(str), monthly_trends[service], marker='o', label=service)

plt.title('Monthly Demand for Services Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Appointments')
plt.xticks(rotation=45)
plt.legend(title='Service Type')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Appointment Status Visualization per Month

```python
# Group by year-month and status, then count frequencies
status_trends = relevant_data.groupby(['year_month', 'status']).size().unstack(fill_value=0)

# Fill any missing status categories with zeros (in case some months do not have all types of statuses)
status_trends = status_trends.reindex(columns=['Completed', 'Pending', 'Cancelled'], fill_value=0)


plt.figure(figsize=(14, 7))
status_trends.plot(kind='area', stacked=True, alpha=0.5, figsize=(14, 7), colormap='viridis')
plt.title('Appointment Status Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Appointments')
plt.xticks(rotation=45)
plt.legend(title='Appointment Status')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns

# Create a copy of the data to encode
encoded_pet_data = pet_data.copy()

# Initialize LabelEncoder
le = LabelEncoder()

# Encoding 'type' and 'gender' to transform them from categorical to numerical
encoded_pet_data['type_encoded'] = le.fit_transform(encoded_pet_data['type'])
encoded_pet_data['gender_encoded'] = le.fit_transform(encoded_pet_data['gender'])

# Calculating the correlation matrix
correlation_matrix = encoded_pet_data[['age', 'weight', 'type_encoded', 'gender_encoded']].corr()

# Plotting the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.05)
plt.title('Correlation Heatmap for Pet Dataset')
plt.show()
```
